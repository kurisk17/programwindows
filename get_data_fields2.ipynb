```python
# 中電実績取得プログラム
# 太陽光の自家消費分は、エリア需要実績や太陽光発電実績には加算していない
# エリア供給力は、特定できるものは電源種別とし、バイオマス混焼や特定できないものは火力に包含(2018年10月1日から揚水欄は揚水発電と揚水動力の合計値）

def Get_data_Chuden():
    chuden_columns = ['Date', 'Time', 'エリア需要', '原子力', '火力', '水力', '地熱', 'バイオマス',
        '太陽光（実績）', '太陽光（出力制御量）', '風力（実績）', '風力（出力制御量）', '揚水', '連系線']
    df_chuden = pd.DataFrame(columns=chuden_columns)
    # nanがないことがわかった
    for i in range(2016,2023):

        df =pd.read_csv('https://powergrid.chuden.co.jp/denki_yoho_content_data/2020_areabalance_current_term.csv',encoding='Shift-JIS',names=list(range(1,15)))
        df.columns=chuden_columns
        df = df[5:]
        df_chuden = pd.concat([df_chuden,df],axis=0)

    df_chuden = df_chuden.reset_index(drop=True)

    df_chuden.to_csv('/Users/yusuke_miyazaki/Desktop/OneDrive/研究関連データ/data_for_project_spot_jikanmae/chuden_jisseki.csv')
```

```python
def Get_data_Kanden():
    year_list = list(range(2016,2023))
    kanden_columns = ['DATE_TIME', 'エリア需要〔MWh〕', '原子力〔MWh〕', '火力〔MWh〕', '水力〔MWh〕','地熱〔MWh〕', 'バイオマス〔MWh〕', '実績〔MWh〕', '抑制量〔MWh〕', '実績〔MWh〕','抑制量〔MWh〕', '揚水〔MWh〕', '連系線〔MWh〕']
    # 空のデータフレームを作る
    df_kanden = pd.DataFrame(columns=kanden_columns)


    for i in year_list:
        df =pd.read_csv(f'https://www.kansai-td.co.jp/denkiyoho/csv/area_jyukyu_jisseki_{i}.csv',encoding='Shift-JIS')
        df.columns= kanden_columns

        df = df.dropna()
        df = df[1:]
        df_kanden = pd.concat([df_kanden,df],axis=0)
    df_kanden= df_kanden.reset_index(drop=True)

    df_kanden.to_csv('/Users/yusuke_miyazaki/Desktop/OneDrive/研究関連データ/data_for_project_spot_jikanmae/kanden_jisseki.csv')

```
