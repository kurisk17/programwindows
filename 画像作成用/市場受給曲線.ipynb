```python
price = np.array(list(range(0,25)))
```

```python
buy = np.cos(np.pi/2*price/price.max())*10
sell = np.sin(np.pi/2*price/price.max())*10
```

```python
sell_sum =[sell[0]]
for i in range(len(sell[1:])):
    sell_sum.append(sell_sum[i] + sell[1+i])
```

```python
buy_sum =[np.sum(buy)]
for i in range(len(buy[1:])):
    buy_sum.append(buy_sum[i] - buy[i])
```

```python
sell_sum =[sell[0]]
for i in range(len(sell[1:])):
    sell_sum.append(sell_sum[i] + sell[1+i])

buy_sum =[np.sum(buy)]
for i in range(len(buy[1:])):
    buy_sum.append(buy_sum[i] - buy[i])

buy_sum=np.array(buy_sum)/10/1.5
sell_sum=np.array(sell_sum)/10/1.5

plt.rcParams['font.family']='MS Mincho'
plt.rcParams['font.size']=20

fig = plt.figure(figsize=(8,8))

ax1 = fig.subplots()
ax2 = ax1.twinx()

ax1.bar(x=price,height=buy,alpha =0.8,color= 'gray',label= '買い入札量')
ax1.bar(x=price,height=sell,alpha =0.6,color= 'gray',label = '売り入札量')
ax1.set_xlabel('Yen/kWh',fontname = 'Times New Roman')
ax1.set_ylabel('容量(MW)')
ax1.axes.xaxis.set_ticks([])
ax1.axes.yaxis.set_ticks([])

# ax1.legend()



# ax2.plot(price,buy_sum,color = 'r',label = '需要曲線')
# ax2.plot(price,sell_sum,color = 'b',label = '供給曲線')
# ax2.axis('off')
# ax1.legend()
# ax2.legend()

ax1.plot(price,buy_sum,color = 'r',label = '需要曲線')
ax1.plot(price,sell_sum,color = 'b',label = '供給曲線')
ax2.axis('off')
ax1.legend()
# ax2.legend()
plt.savefig('../../../卒論関連書類/画像/説明図/シングルプライス説明.png')
plt.show()
```
